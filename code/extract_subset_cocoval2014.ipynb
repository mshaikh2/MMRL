```python
from pycocotools.coco import COCO
import os, sys
import json
import numpy as np
from shutil import copyfile
from tqdm import tqdm
```

```python
coco_dir = '/media/MyDataStor1/mmrl/MMRL/data/coco'
ann_dir = os.path.join(coco_dir, 'annotations', 'instances_val2014.json')
img_dir = os.path.join(coco_dir, 'images')
```

```python
coco = COCO(ann_dir)
cd = coco.dataset
```

```python
# get id number of label of interest
class_name = 'teddy bear'
cc = coco.getCatIds(catNms=[class_name])
print(cc)
```

```python
# get all the image ids containing the given label in val2014 set
cid = coco.getImgIds(catIds=[88])
len(cid)
```

```python
# get all the image elements and filenames
rst = coco.loadImgs(ids=cid)
len(rst)
```

```python
urls = [p['file_name'] for p in rst[:100]]
```

```python
tar_dir = '/media/MyDataStor1/mmrl/MMRL/data/rep_eval'
for d in tqdm(urls):
    src = os.path.join(img_dir, d)
    dst = os.path.join(tar_dir, class_name, d)
    copyfile(src, dst)
```
