```python
cfg_from_file('cfg/DAMSM/coco.yml')

imsize = cfg.TREE.BASE_SIZE * (2 ** (cfg.TREE.BRANCH_NUM-1))
batch_size = cfg.TRAIN.BATCH_SIZE
image_transform = transforms.Compose([
    transforms.Scale(int(imsize * 76 / 64)),
    transforms.RandomCrop(imsize),
    transforms.RandomHorizontalFlip()])
dataset = TextDataset(cfg.DATA_DIR, 'train',
                      base_size=cfg.TREE.BASE_SIZE,
                      transform=image_transform)
print(dataset.n_words, dataset.embeddings_num)
dataloader = torch.utils.data.DataLoader(
    dataset, batch_size=batch_size, drop_last=True,
    shuffle=False, num_workers=0)
```

```python
d = iter(dataloader).next()
imgs, caps, cap_len, cls_id, key, cap_imgs, cap_imasks, sent, sent_mask = prepare_data(d)
```

```python
f_tokens = np.zeros((fake_tokens.shape[0], 15), dtype=np.int64)
f_cap_lens = []
cnt=0
for i in fake_tokens: 
    temp = np.array([x for x in i if x!=27299 and x!=0])
    num_words = len(temp)
    if num_words <= 15:
        f_tokens[cnt][:num_words] = temp
    else:
        ix = list(np.arange(num_words))  # 1, 2, 3,..., maxNum
        np.random.shuffle(ix)
        ix = ix[:15]
        ix = np.sort(ix)
        f_tokens[cnt] = temp[ix]
        num_words = 15
    f_cap_lens.append(num_words)
    cnt += 1
```

```python
import torch
from model import TEXT_TRANSFORMER_ENCODERv2, CNN_ENCODER
from miscc.config import cfg
```

```python
image_encoder = CNN_ENCODER(cfg.TEXT.EMBEDDING_DIM)
```

```python
image_encoder
```

```python
frozen_list_image_encoder = ['Conv2d_1a_3x3','Conv2d_2a_3x3','Conv2d_2b_3x3','Conv2d_3b_1x1','Conv2d_4a_3x3']

for p in image_encoder.parameters(): # make image encoder grad on
    p.requires_grad = True
for k,v in image_encoder.named_children(): # freeze the layer1-5 (set eval for BNlayer)
    if k in frozen_list_image_encoder:
        v.train(False)
        v.requires_grad_(False)
```

```python
# check parameters in the model
for n, p in image_encoder.named_parameters(): # make image encoder grad on
    print(n, ': ', p.requires_grad)
```

```python
import torch.nn as nn
xi = torch.tensor([[1,2,3]], dtype=torch.float, requires_grad=True)
li = nn.Linear(3,3)
li.requires_grad_()
lo1 = nn.Linear(3,3)
lo1.requires_grad_()
lo2 = nn.Linear(3,3)
lo2.requires_grad_()
```

```python
xf = li(xi)
print(xf)
xo1 = lo1(xf)
xo2 = lo2(xf)
print(xo1)
print(xo2)
```

```python
ls1 = xo1 * 2
ls2 = xo2 * 5
li.zero_grad()
lo1.zero_grad()
lo2.zero_grad()
ls1.backward()
print('ls1 backpropagation sucess!')
ls2.backward()
print('ls2 backpropagation sucess!')
```
